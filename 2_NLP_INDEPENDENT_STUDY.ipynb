```python
def data_cleaning(df):
    df['ANSWER'] = df['ANSWER'].str.replace('\n', '')
    df['ANSWER'] = df['ANSWER'].str.replace("'", '')
    df['ANSWER'] = df['ANSWER'].str.replace('-', '')
    df['ANSWER'] = df['ANSWER'].str.replace(',', '')
    df['ANSWER'] = df['ANSWER'].str.replace(':', '')
    df['ANSWER'] = df['ANSWER'].str.replace('\d+', '', regex=True)    
```

```python
import spacy

def removing_stop_words(df):
    # Load the spaCy English language model
    nlp = spacy.load("en_core_web_sm")

    # Create a new column to store the filtered words
    df['FILTERED_WORDS'] = ''

    # Iterate over the rows of the DataFrame
    for index, row in df.iterrows():
        text = row['ANSWER']
        # Process the text using spaCy
        doc = nlp(text)
        # Remove stopwords
        filtered_words = [token.text for token in doc if not token.is_stop]
        # Store the filtered words in the new column
        df.at[index, 'FILTERED_WORDS'] = ' '.join(filtered_words)

    return df
```

#### CODE TO SPLIT SENTENCES FROM DATA FRAME AND STORE IT IN A LIST 

```python
def splitting_sentences(df):
    # Create a new column to store the filtered words
    splitted_words = np.array()

    # Iterate over the rows of the DataFrame
    for index, row in df.iterrows():
        text = row['FILTERED_WORDS']
        split_words = re.split(r'[.!]', text)
        # Store the filtered words in the new column
        splitted_words.append(split_words)

    final_new_df1 = []
    for ele in splitted_words:
        for sentence in ele:
            if sentence.strip() != '':
                final_new_df1.append(sentence.strip())

    return final_new_df1
```
